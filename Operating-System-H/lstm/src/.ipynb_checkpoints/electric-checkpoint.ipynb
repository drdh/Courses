```python
import numpy as np
import matplotlib.pyplot as plt
import os
import re
```

```python
fld=open(os.path.join("../data","LD2011_2014.txt"),"r")
data=[]
cid=250
for line in fld:
    if line.startwith('"";'):
        continue
    cols=[float(re.sub(",",".",x)) for x in line.strip().split(";")[1:]]
    data.append(cols[cid])
fld.close()
```
